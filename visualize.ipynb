### Visualization

### 1. Walking Speed Analysis

### Scatter Plot of Age vs. Walking Speed

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
ms_df = pd.read_csv('ms_data_new.csv')
sns.lmplot(data=ms_df, x='age', y='walking_speed', line_kws={'color':'red'})
plt.title("Age vs. Walking Speed")
plt.show()
```

### Box Plots by Education Level

```python
sns.boxplot(data=ms_df, x='education_level', y='walking_speed')
plt.title("Walking Speed by Education Level")
plt.show()
```

### Line plot of education & age interaction

```python
sns.lineplot(data=ms_df, x='age', y='walking_speed', hue='education_level', style='education_level')
plt.title("Education/Age Interaction Effect on Walking Speed")
plt.show()
```

## 2. Cost Analysis

### Bar Plot: Mean Costs by Insurance Type

```python
sns.barplot(data=ms_df, x='insurance_type', y='visit_cost')
plt.title("Mean Costs by Type of Insurance")
plt.show()
```

### Box Plots: Cost Distributions

```python
sns.boxplot(data=ms_df, x='insurance_type', y='visit_cost')
plt.title("Distribution of Costs by Insurance Type")
plt.show()
```

## 3. Combined Visualizations

```python
sns.pairplot(data=ms_df)
plt.title("Pairplots of Key Variables")
plt.show()
```

### Faceted plot by Education/Insurance

```python
plot = sns.FacetGrid(ms_df, col='education_level')
plot.map(sns.histplot, 'insurance_type')
```

### Time trends

```python
from datetime import datetime, timedelta
plt.figure(figsize=(12,6))
sns.lineplot(data=ms_df, x='visit_date', y='walking_speed', marker='o')
plt.xlabel("Visit Date")
plt.ylabel("Walking Speed")
plt.title("Time Trends: Walking Speed over Multiple Visit Dates")
plt.show()
```
